```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
from IPython.display import display

# Load a pre-trained Stable Diffusion model from Hugging Face
def load_text_to_image_model():
    model_id = "CompVis/stable-diffusion-v1-4"  # Replace with the desired model ID
    pipeline = StableDiffusionPipeline.from_pretrained(model_id)
    pipeline = pipeline.to("cuda" if torch.cuda.is_available() else "cpu")  # Use GPU if available
    return pipeline

# Generate an image from text prompt
def generate_image_from_text(pipeline, prompt):
    image = pipeline(prompt).images[0]  # Generate image
    return image

# Main function to run the pipeline
def main():
    # Step 1: Input prompt from user
    prompt = input("Enter your prompt: ")

    # Step 2: Load model
    pipeline = load_text_to_image_model()

    # Step 3: Generate image
    print("Generating image...")
    generated_image = generate_image_from_text(pipeline, prompt)

    # Save the generated image
    generated_image.save("generated_image.png")

    # Step 4: Display the generated image in Colab
    print("Displaying generated image in Colab...")
    display(generated_image)  # Use IPython display to show the image

# Run the pipeline
if __name__ == "__main__":
    main()
```
